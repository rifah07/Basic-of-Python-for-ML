```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#normal Distribution;
#loc = mean; scale= standard deviation
#size= shape/dimantion(s) of returned array 
data= np.random.normal(loc=0, scale=1, size=100)

#bins specifies the number of bins (intervals) to divide the data into.
#In this case, the histogram will have 30 bins
plt.hist(data, bins=30, edgecolor='black')
plt.title("Histogram of Data Distribution")
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
mean_value= np.mean(data)
median_value= np.median(data)
std_dev= np.std(data)

print(f"Mean of the data is: {mean_value}")
print(f"Median of the data is: {median_value}")
print(f"Standard Deviation of the data is: {std_dev}")
```
