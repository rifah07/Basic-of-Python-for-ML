```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris= load_iris()
```

```python
#view data description and information of the dataset
print(iris.DESCR)
```

```python
iris
```

Feature Data

```python
df_iris= pd.DataFrame(iris.data)
df_iris
```

```python
df_iris.columns= iris.feature_names
df_iris.head()
```

```python
target= pd.DataFrame(iris.target)
target.head()
```

```python
#0 is Iris Setosa; 1 is Iris Versicolour; 2 is Iris Virginica

target = target.rename(columns = {0: 'target (variety)'})
target.tail()
```

Exploratory Data Analysis (EDA)

```python
##in the code above, axis = 1 tells the data frame to add the target data frame
#as another column of the data data frame,
#axis = 0 would add the values as another row on the bottom

df_full_iris= pd.concat([df_iris , target], axis=1)

df_full_iris.tail()
```

Petal length and width is most correlated with the target, meaning that as these numbers increase, so does the target value. Modeling

```python
df_full_iris
```

```python
X= df_full_iris.iloc[:, :4] #X value is all the input factors/predictors/feature data/all except the last column
X
```

```python
Y= df_full_iris.iloc[:, 4] #Y is the output or target data, the last column
Y
```

Standardize

```python
scaler= StandardScaler()
X= scaler.fit_transform(X)
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(X,Y, test_size=0.2, random_state=1, stratify= Y)

#by stratifying on y we assure that the different classes are represented proportionally to the amount 
#in the total data (this makes sure that all of class 1 is not in the test group only)
```

Baseline Prediction

The baseline is the probability of predicting class before the model is implemented. Here we can see the ratio the class in target column. The goal of our model is to improve on this baseline, or random prediction. Also, if there is a strong class imbalance (if 90% of the data was in class 1), then we could alter the proportion of each class to help the model predict more accurately.

```python
df_full_iris.iloc[:, 4].value_counts(normalize=True) # or Y.value_counts(normalize=True)
```

Here 3 types/classes are equaly distributed. The baseline prediction for this model is 1/3.

Logistic Regression Model

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
y_predict= model.predict(X_test)
```

```python
#the testing data, or X_test, is necessary for evaluating a model's performance because 
#it provides unseen data for the model to make predictions on. Here's how it works:
#Predicting: The model makes predictions on the testing data X_test.
#Comparing: These predictions are then compared to the actual labels y_test to see how well the model did.
#The model.score(X_test, y_test) method simplifies this process by internally making predictions on X_test
#and then comparing them to y_test to calculate the performance metric.
#In short, you need the testing data to generate the predictions, and you compare those predictions with
#the actual labels to evaluate the model's performance accurately.

model.score(X_test, Y_test)
```

```python
# The test score is already very high, but we can use the cross validated score to ensure the model's strength 

score= cross_val_score(model, X_train, Y_train, cv=10)

#Cross-Validation (cross_val_score): This function comes from the scikit-learn library and 
#is used to evaluate a model by splitting the data into several subsets, or â€œfolds.â€
#Parameters:
#model: The machine learning model you want to evaluate.
#X_train: The feature data used for training.
#y_train: The target labels used for training.
#cv=10: This specifies 10-fold cross-validation. It means that the data is divided into 10 subsets.
#The model is trained on 9 of these subsets and tested on the remaining subset.
#This process is repeated 10 times, each time with a different subset used for testing.
#Output (scores): This variable will store the performance score of the model for each fold.
#Typically, itâ€™s an array of scores, one for each fold.

average_score= score.mean()
print(average_score)
```

Without any adjustments or tuning, this model is already performing very well with a test score of .9667 and a cross validation score of .9499. This means that the model is predicting the correct class for the flower about 95% of time. Much higher than the baseline of 33%!


Let's Understand the Predictions

```python
#Coefficients

coefficient= model.coef_
df_coff= pd.DataFrame(coefficient, columns=iris.feature_names)
df_coff
```

We can get an idea of how much of an impact each of the features had in deciding if a flower belonged to that class. For instance, petal length was barely a deciding factor for if a flower was in class 1, but petal width was a strong predictor for class 2.

Predicted Values

```python
comparison_df= pd.DataFrame({'actual': Y_test, 'predicted': y_predict})
comparison_df= comparison_df.reset_index(drop=True)
comparison_df
```

Confusion Metrics

```python
conf= confusion_matrix(Y_test, y_predict)
conf_met= pd.DataFrame(conf)
conf_met
```

We can see class 2 as class 1 in 1 instance is incorrectly labeled. But al others are correct.

```python
#Classification Report

cr = classification_report(Y_test, y_predict)
print(cr)
```

```python
# Predicted Probabilities

pp= model.predict_proba(X_test)

Y_pp= pd.DataFrame(pp, columns=['class0_pp','class1_pp','class2_pp']) #classes means traget(varity) class
Y_pp.tail()
```
