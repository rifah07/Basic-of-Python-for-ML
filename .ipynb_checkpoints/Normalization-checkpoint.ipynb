```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
```

```python
df= pd.read_csv('data.csv')
X= df[['Volume','Weight']]
X
```

```python
scale_mnmx= MinMaxScaler()
```

```python
sacledX1 = scale_mnmx.fit_transform(X)
print(sacledX1)
```

Standard Scaler
Standardization scales data to have a mean of 0 and a standard deviation of 1.

```python
scale_stds= StandardScaler()
```

```python
scaledX2= scale_stds.fit_transform(X)
print(scaledX2)
```

Robust Scaling
Robust Scaling uses the median and the interquartile range to scale the data, making it robust to outliers.

```python
scale_rbs = RobustScaler()
```

```python
scaledX3= scale_rbs.fit_transform(X)
print(scaledX3)
```
