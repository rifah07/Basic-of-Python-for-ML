```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(42)
#The use of the number 42 as a seed value is mostly a fun and quirky tradition in the coding community.
#Its popularity stems from the book "The Hitchhiker's Guide to the Galaxy" by Douglas Adams,
#where 42 is humorously presented as "the answer to the ultimate question of life, the universe, and everything."
#Because of this cultural reference, many programmers use 42 as a seed value as an inside joke or a playful nod to the book.
#However, any integer value can be used as a seed, and it will serve the same purpose of creating
#a reproducible sequence of random numbers. The choice of 42 is not inherently better or worse than any other number;
#it’s just a bit of a tradition and a way to have some fun with code.
```

```python
X= 6 * np.random.rand(100,1) -3
Y= 0.5 * X**2 + X + 2 + np.random.randn(100,1)
```

```python
polynomial_features= PolynomialFeatures(degree=2, include_bias=False)
X_poly= polynomial_features.fit_transform(X)
```

```python
poly_regression=LinearRegression()
poly_regression.fit(X_poly, Y)
```

```python
X_new= np.linspace(-3,3,100).reshape(100,1)
X_new_poly= polynomial_features.transform(X_new)
Y_new= poly_regression.predict(X_new_poly)
```

```python
plt.scatter(X,Y, alpha=0.7, label='Original Data')
plt.plot(X_new,Y_new, color='red', linewidth=2,label='Polynomial Regression')
plt.title('Polynomial Regression by Python')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
