```python
from sklearn.datasets import load_iris
import seaborn as sns
import pandas as pd
```

```python
iris= load_iris()
```

```python
#view data description and information of the dataset
print(iris.DESCR)
```

```python
iris
```

Feature Data

```python
df_iris= pd.DataFrame(iris.data)
df_iris
```

```python
df_iris.columns= iris.feature_names
df_iris.head()
```

```python
target= pd.DataFrame(iris.target)
target.head()
```

```python
#0 is Iris Setosa; 1 is Iris Versicolour; 2 is Iris Virginica

target = target.rename(columns = {0: 'target (variety)'})
target.tail()
```

Exploratory Data Analysis (EDA)

```python
##in the code above, axis = 1 tells the data frame to add the target data frame
#as another column of the data data frame,
#axis = 0 would add the values as another row on the bottom

df_full_iris= pd.concat([df_iris , target], axis=1)

df_full_iris.tail()
```

Data Cleaning

```python
df_full_iris.dtypes
```

```python
#checking missing values

df_full_iris.isnull().sum()
```

```python
#statistival overview

df_full_iris.describe()
```

Visualizing

```python
##annot = True adds the numbers onto the squares

sns.heatmap(df_full_iris.corr(), annot=True)
```
