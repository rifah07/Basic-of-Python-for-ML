```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Normal distributed data
exam_scores= np.random.normal(loc=60, scale=10, size=100)
```

```python
plt.hist(exam_scores, bins=30, edgecolor='black')
plt.title('Histogram for Exam Scores')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.show()
```
