This is an approach of Unsupervised learning. So no labeling is needed, no trainkng, testing is also needed.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
```

```python
x= [2,1,4,6,11,23,100,20,12,24,200,50,60,18]
y= [20,12,43,60,101,203,120,28,120,204,209,58,66,180]

plt.scatter(x,y)
plt.show()
```

```python
x_y= list(zip(x,y))

print(x_y)
```

```python
linkage_data= linkage(x_y, method='ward', metric='euclidean')
dendrogram(linkage_data)

plt.show()
```

```python
hierarchical_cluster = AgglomerativeClustering(n_clusters=2, linkage='ward')
labels= hierarchical_cluster.fit_predict(x_y)

labels
```

```python
plt.scatter(x,y, c=labels)
plt.show()
```

Let us use a Dataset

```python
from sklearn.datasets import make_blobs
```

```python
X, Y= make_blobs(n_samples= 500, centers= 7, cluster_std= 0.70, random_state=0)
```

```python
agg_clus= AgglomerativeClustering(n_clusters=5)
y_pred= agg_clus.fit_predict(X)
```

Scatter Diagram

```python
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='Pastel1') #cmap= Set1/Dark2/BrBG/Purples/Blues/viridis/plasma/inferno
plt.title("Aggromative Clustering")
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

Plot the Dendrogram

```python
# generate the linkage matrix
lnk= linkage(X, method='ward')

#plt.figure(figsize=(20, 10))
dendrogram(lnk)

plt.title("Dendogram")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
plt.show()
```
